### **Experimento 1**

Tenemos 1'000.000 de dolares que invertimos hace 2 años, esta cantidad es distribuida en partes iguales a 5 empresas, es decir, la cantidad de acciones que se compraran por empresa sera por un valor 200.000 dolares

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2019, 5, 4)  # Set Start Date
        self.SetEndDate(2021, 5, 4) #Set End Date
        self.SetCash(1000000)  # Set Strategy Cash
        
        
        self.nflx = self.AddEquity("FDX", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.tsla = self.AddEquity("ZM", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.amzn = self.AddEquity("LEVI", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.goog = self.AddEquity("NVDA", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.aapl = self.AddEquity("AMZN", Resolution.Minute)
        self.aapl.SetDataNormalizationMode(DataNormalizationMode.Raw)
            
    def OnData(self, data):

        if not self.Portfolio.Invested:
            self.SetHoldings("FDX", 0.2)
            self.SetHoldings("ZM", 0.2)
            self.SetHoldings("LEVI", 0.2)
            self.SetHoldings("NVDA", 0.2)
            self.SetHoldings("AMZN", 0.2)
            
            
            self.Log("Active Pos Fdex " + str(self.Portfolio["FDX"].Quantity))
            self.Log("Active Pos Zoom " + str(self.Portfolio["ZM"].Quantity))
            self.Log("Active Pos Levi " + str(self.Portfolio["LEVI"].Quantity))
            self.Log("Active Pos Nvidia " + str(self.Portfolio["NVDA"].Quantity))
            self.Log("Active Pos Amazon " + str(self.Portfolio["AMZN"].Quantity))
```
